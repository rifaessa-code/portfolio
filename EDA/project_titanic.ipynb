```python
# import library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#show dataset
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
# 3. melihat struktur data
titanic.shape
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
#4. Cek missing values

titanic.isnull().sum()
```

```python
#visualisasikan

sns.heatmap(titanic.isnull(), cbar=False)
plt.show()
```

```python
#5. cek duplikasi data

titanic.duplicated().sum()
```

```python
#6. Histogram usia penumpang
plt.figure(figsize=(8,5))
plt.hist(titanic['Age'].dropna(), bins=30)
plt.title('Distribusi Usia Penumpang')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```

```python
# boxplot untuk melihat outlier usia
plt.figure(figsize=(6,4))
sns.boxplot(x=titanic['Age'])
plt.title('Boxplot Usia')
plt.show()
```

```python
#7. visualisasi jenis kelamin
titanic['Sex'].value_counts().plot(kind='bar')
plt.title('Jumlah Penumpang per Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Jumlah')
plt.show()
```

```python
#8. distribusi kelas penumpang

sns.countplot(x='Pclass', data=titanic)
plt.title('Distribusi Kelas Penumpang')
plt.show()
```

```python
# survived rate by sex

sns.countplot(x='Survived', hue='Sex', data=titanic)
plt.title('Survived passengers by sex')
plt.show()
```

```python
# survival rate by passengers class

sns.countplot(x='Survived', hue='Pclass', data=titanic)
plt.title('Survived based on passengers class')
plt.show()
```

```python
# Survival rate by embarked

sns.countplot(x='Embarked', hue='Survived', data=titanic)
plt.title('Survived Passengers by Embarked')
plt.xlabel('embarked')
plt.ylabel('amount of passengers')
plt.show()
```

```python
#Cross tabulation

pd.crosstab([titanic['Sex'], titanic['Embarked']], titanic['Survived'])
```
